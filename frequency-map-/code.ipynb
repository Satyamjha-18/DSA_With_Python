### frequency map : store frequency in dictionary

```python
n = [5,6,4,8,3,11,5,4,3,7,5,1,5,1,1]
```

```python
freq_map=dict()

for i in range(0,len(n)):
    if n[i] in freq_map:
        freq_map[n[i]]+=1
    else:
        freq_map[n[i]]=1

print(freq_map)

```

```python
num = "122334455"

freq_map = {}

for i in range(len(num)):
    if num[i] in freq_map:
        freq_map[num[i]] += 1
    else:
        freq_map[num[i]] = 1

print(freq_map)
```
