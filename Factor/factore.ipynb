###  Factor of any number 

##### Method:01

```python
num = 20

result =[]

for i in range (1,num+1):
    if num %i==0:
        result.append(i)

print(result)
```

```python
num = int(input("Enter the number :"))

factor =[]

for i in range(1,num+1):
    if num %i==0:
        factor.append(i)

print("Factor of ",num,"=",factor)
```

##### Method 02 : Better solution

```python
num = 20

result=[]
for  i in range(1,num//2):
    if num %i==0:
        result.append(i)

result.append(num)
print("Factor of ",num,"=",result)
```

### Method 03: optimal solution 

```python
num = 20 
from math import sqrt

result=[]

for i in range(1,int(sqrt(num)+1)):
    if num %i==0:
        result.append(i)

    if num //i!= i:
        result.append(num//i)
print(result)
```
