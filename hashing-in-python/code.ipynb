### With Hashing

```python
n=[1,2,1,3,2,1,5]
```

```python
hash_list=[0]*8

for i in n :
    hash_list[i]+=1
print(hash_list)
```

```python
arr = [1, 1, 1, 2, 2, 3, 3, 3, 3, 4, 4]

# Step 1: Frequency Map
hash_map = {}

for i in arr:
    hash_map[i] = hash_map.get(i, 0) + 1

print("Frequency Map:", hash_map)

# Step 2: Find Highest and Second Highest Frequency
highest = 0
second = 0

for key, value in hash_map.items():

    if value > highest:
        second = highest
        highest = value

    elif value > second and value < highest:
        second = value

print("Highest Frequency:", highest)
print("Second Highest Frequency:", second)

# Step 3: Print the element having second highest frequency
for key, value in hash_map.items():
    if value == second:
        print("Second Highest Frequency Element:", key)
        break
```

#### Multiple Queries

```python
arr = [1,2,2,3,4,4,4,5]

frq={}

for i in arr:
    frq[i]=frq.get(i,0)+1

queries=[2,4,6]

for q in queries:
    print(q,"->",frq.get(q,0))
```

```python
n = [5,3,2,2,1,5,5,7,5,10]
m=[10,111,1,9,5,67,2]
```

### brute solution 

```python
for num in m:
    count = 0          # Har query ke liye reset
    for x in n:
        if x == num:
            count += 1

    print(num, "->", count)
```

### method 2:

```python
hash_list=[0]*11

for num in n:
    hash_list[num]+=1

for num in m:
    if num <1 or num >10:
        print(0)

    else:
        print(hash_list[num])
```

```python
n = [5, 3, 2, 2, 1, 5, 5, 7, 5, 10]
m = [10, 111, 1, 9, 5, 67, 2]

# Step 1: Create Frequency Map
hash_map = {}

for num in n:
    hash_map[num] = hash_map.get(num, 0) + 1

print("Frequency Map:", hash_map)

# Step 2: Answer Queries
for query in m:
    print(query, "->", hash_map.get(query, 0))
```
